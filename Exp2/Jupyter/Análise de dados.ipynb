# Parte B

Nesta parte do trabalho apenas ligamos a clula fotoeltrica a um condensador e medimos o potencial nos terminais deste. Esta  outra forma de determinar o potencial de paragem pois os electes ejetados pela luz carregam o condensador at este atingir a tenso de paragem, onde o campo eltrico do condensador impede novos electres de se acumularem. Essa tenso  medida no multmetro.

Como podemos ver pelo grfico acima, obtivemos ajustes ligeiramente diferentes da parte A devido a fugas de carga no condensador, resistncia interna da clula fotoeltrica e recolha incompleta dos eletres ejetados, o que reduz a tenso acumulada. \
Com este mtodo obtivemos os seguintes valores: \
Coeficiente Angular (h/e): 3.568e-15 $Vs$\
Constante de Planck (h): 5.715e-34 $Js$\
Intercept (-$\phi$/e): -1.558e+00 $V$\
Funo Trabalho ($\phi$): 2.496e-19 $J$\
Coeficiente de Determinao (R): 0.9630\
Incerteza Relativa: 6.93%

# Concluses

Este estudo permitiu-nos verificar o efeito fotoeltrico e a dependncia linear da frequncia dos leds com a energia cintica de dos eletres ejetados. Isto deu-nos tambm uma forma de direta de calcular a constante de Planck. \
Fizemos este estudo utilizando dois mtodos diferentes onde podemos estudar as mais valias de cada um dos mtodos e, apesar de uma incerteza relativa maior no mtodo do condensador face  que obtemos no ajuste da parte A, considerar a propagao de incertezas dos ajustes dos potenciais de paragem de cada um dos leds  e a incerteza da tenso de paragem vs frequncia leva-nos a concluir que o mtodo da parte B  mais preciso simplesmente porque estamos a ler correntes muito pequenas na parte A.\
Este  um trabalho que incorre em muitos erros sistemticos pois as condies no so ideais (material desgastado impossibilitando a cobertura total da clula passando luz ambiente), os valores que estamos a ler so muito pequenos e qualquer flutuao ou interferncia externa pode afetar profundamente as leituras.
